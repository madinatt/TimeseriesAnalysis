```python
import matplotlib.pyplot as plt

date_rng = pd.date_range(start = "2020-03-07", end = "2020-03-16", freq = "D")
df = pd.DataFrame(date_rng, columns = ['date'])
df['data'] = [1,3, 4, 6,4,5,8,7,9,11]
```

```python
plt.figure(figsize=(8,4))
plt.plot(df["date"], df["data"], marker = "o", linestyle = "-")
plt.xlabel("Date")
plt.ylabel("Data Value")
plt.title("Simple Univariate Time Series")
```

```python
np.random.seed(1)
date_rng = pd.date_range(start = "2020-03-07", end = "2020-03-16", freq = "D")
temparature = np.random.randint(20,30, size = (len(date_rng),))
humidity = np.random.randint(40,60, size = (len(date_rng),))
```

```python
df = pd.DataFrame(date_rng, columns = ['date'])

df
```

```python
df["temparature"] = temparature
df["humidity"] = humidity

df
```

```python
plt.figure(figsize=(12,6))

plt.subplot(2,1,1)
plt.plot(df["date"], df["temparature"], marker = "o", label = "Temparature")
plt.xlabel("Date")
plt.ylabel("Temparature")
plt.title("Temparature over time")

plt.subplot(2,1,2)
plt.plot(df["date"], df["humidity"], marker = "o", label = "Humidity", color = "orange")
plt.xlabel("Date")
plt.ylabel("Humidity")
plt.title("Humidity over time")

plt.tight_layout()
```
